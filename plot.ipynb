```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.size'] = 30
# plt.rcParams['font.weight'] = 'bold'
# plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['lines.linewidth'] = 5
plt.rcParams['lines.markersize'] = 10
```

```python
# filter = 5

test_errors = np.loadtxt('result/2TX-20-min-dist/test_errors.txt')
test_misses = np.loadtxt('result/2TX-20-min-dist/test_misses.txt')
test_falses = np.loadtxt('result/2TX-20-min-dist/test_falses.txt')

fit, axs = plt.subplots(1, 3, figsize=(30, 8))
axs[0].plot(test_errors[1:, 0])
axs[0].set_title('2TX Localize error')
axs[0].set_ylabel('1 cell is 10m')
axs[0].set_xticks(range(9))
axs[0].set_xlabel('epoch')

axs[1].plot(test_misses[1:, 0])
axs[1].set_xticks(range(9))
axs[1].set_title('Miss counts')
axs[1].set_xlabel('epoch')

axs[2].plot(test_falses[1:, 0])
axs[2].set_title('False counts')
axs[2].set_xticks(range(9))
axs[2].set_xlabel('epoch')
```
